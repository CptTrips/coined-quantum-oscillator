The displacement, $d$, should be larger than the thermal width of the particle $\Delta x$
$$
    d \geq \Delta x
$$

Using $d = \frac{2 g \mu_B \partial_z B}{m \omega^2}$ and $\Delta x = \sqrt{\frac{\hbar}{2m\omega} \coth(\frac{\beta \hbar \omega}{2})}$

$\beta \hbar \omega = \ln(1 + 1/n)$

$\coth(\beta \hbar \omega / 2) \approx 2/\beta\hbar\omega$ when $\beta \hbar \omega/2 << 1$ and $\beta \hbar \omega \approx 1/n$ therefore
$$
    \coth(\frac{\beta \hbar \omega}{2}) \approx 2 n
$$

$$
    \frac{2 g \mu_B (\partial_z B)}{m \omega^2} \geq \sqrt{\frac{2\hbar}{m \omega} n} \\
$$

$$
    \partial_z B \geq \sqrt{\Big(\frac{ \hbar}{2 g^2 \mu_B^2}\Big) m \omega^3 n}
$$

```python
import numpy as np
from scipy.constants import hbar, pi
from matplotlib import pyplot as plt

%matplotlib notebook

g = 2
mu_B = 9.27e-24

k = hbar / (2 * g**2 * mu_B**2)

def min_gradient(mw3, n):
    
    return np.sqrt(k * mw3 * n)

```

```python
n = np.linspace(1, 100, 100)

f_array = np.arange(2,6) * 2.5e4

mass = (4/3) * pi * (25e-9)**3 * 3510

plt.figure()

for f in f_array:
    
    mw3 = mass * (2 * pi * f)**3
        
    plt.plot(n, min_gradient(mw3, n)/1e6, label=('$\omega = 2 \pi$ {:.3g} kHz'.format(f/1000)))
    
plt.ylim(0,1)

plt.xlim(0, 75)
    
plt.xlabel('$\\bar{n}$')

plt.ylabel('$\partial_z B$ (T/$\mu$m)')
             
plt.legend()

plt.show()

```

```python
omega = 2 * np.pi * 6.5e4
print(min_gradient(mass*omega**3, 20)/1e6, " T/m 10^6")
```
